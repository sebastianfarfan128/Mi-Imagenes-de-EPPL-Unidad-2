"Como estudiante de preparatoria, necesitas desarrollar un programa en Python que funcione como una calculadora geométrica básica. El programa debe mostrar un menú interactivo que permita al usuario elegir entre tres operaciones: calcular el área de un círculo, calcular su perímetro (circunferencia) o calcular el volumen de una esfera. El menú debe repetirse continuamente hasta que el usuario seleccione la opción de salir. El código debe estar organizado en funciones específicas y utilizar estructuras de control (if, elif, else, while) para validar la entrada del usuario y dirigir el flujo del programa."
1. Área del círculo
A = π × r²
π (pi) multiplicado por el radio al cuadrado
2. Perímetro (Circunferencia)
P = 2 × π × r
2 veces π por el radio
3. Volumen (Esfera)
V = (4/3) × π × r³
4 tercios de π por el radio al cubo

![texto del vinculo](https://raw.githubusercontent.com/sebastianfarfan128/Mi-Imagenes-de-EPPL-Unidad-2/refs/heads/main/circulo.PNG)

```python
import math
# Elaborado por Sebastian Farfan
# 1. Función para calcular el área
def areacirculo():
    print("\n--- Cálculo de Área ---")
    radio = float(input("Ingresa el radio del círculo: "))
    area = math.pi * (radio ** 2)
    print(f"✅ El área del círculo es: {area:.2f} unidades²")

# 2. Función para calcular el perímetro
def perimetrocirculo():
    print("\n--- Cálculo de Perímetro ---")
    radio = float(input("Ingresa el radio del círculo: "))
    perimetro = 2 * math.pi * radio
    print(f"✅ El perímetro del círculo es: {perimetro:.2f} unidades")

# 3. Función para calcular el volumen (esfera)
def volumencirculo():
    print("\n--- Cálculo de Volumen (Esfera) ---")
    radio = float(input("Ingresa el radio de la esfera: "))
    volumen = (4/3) * math.pi * (radio ** 3)
    print(f"✅ El volumen de la esfera es: {volumen:.2f} unidades³")

# 4. Función del menú principal
def menu():
    # 'while True' mantiene el programa corriendo hasta que el usuario quiera salir
    while True:
        print("\n" + "="*30)
        print("        MENÚ PRINCIPAL")
        print("="*30)
        print("1. Calcular área del círculo")
        print("2. Calcular perímetro del círculo")
        print("3. Calcular volumen (esfera)")
        print("4. Salir")
        print("="*30)

        opcion = input("Selecciona una opción (1-4): ")

        # Uso de if, elif y else para controlar el flujo
        if opcion == '1':
            areacirculo()               # Llamada a función
        elif opcion == '2':
            perimetrocirculo()          # Llamada a función
        elif opcion == '3':
            volumencirculo()            # Llamada a función
        elif opcion == '4':
            print("👋 ¡Gracias por usar el programa. Hasta luego!")
            break                       # Rompe el bucle while y termina el programa
        else:
            print("⚠️ Opción no válida. Por favor, ingresa un número del 1 al 4.")

# Punto de entrada del programa
if __name__ == "__main__":
    menu()
```
